````python
# histogramă

# bibliotecă pentru calcule numerice
import numpy as np
# biblioteca pentru grafice
import matplotlib.pyplot as plt

# închide toate graficele anterioare (dacă există)
plt.close("all")

# alege aleatoriu și uniform 1000 numere
ur = np.random.rand(1000)

# desenează ca histogramă
plt.hist(ur)
# pe axa OX vrem această etichetă (legendă)
plt.xlabel("Numere aleatorii distribuite uniform")

# arată graficul
plt.show()
````

Histograma poate fi poziționată și pe orizontală.

````python
# histogramă orizontală
import matplotlib.pyplot as plt
import numpy as np

# inițializăm o "sămînță" pentru generarea numerelor aleatorii,
# lucru care asigură că se vor alege valori similare la rerulare
np.random.seed(21333111)

plt.rcdefaults() # revino la setările standard, dacă au existat modificări
fig, ax = plt.subplots() # creează o figură pe care o denumim 'fig', iar 'ax' va fi graficul din ea

# date ca exemplu
people = ('Ion', 'Maria', 'George', 'Dragoș', 'Vasile', 'Oana')
y_pos = np.arange(len(people))
performance = 3 + 10*np.random.rand(len(people)) # generează valori obținute de persoane
error = np.random.rand(len(people)) # genereazăși erori

ax.barh(y_pos, performance, xerr=error, align='center') # pune barele orizontal
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis()
ax.set_xlabel('Performance')
ax.set_title('Cîte gogoși ai mîncat azi?')

plt.show()
````

Acum un exemplu de tip "scatter plot", cu tot cu curba teoretică.

````python
# scatter
import numpy as np 
import matplotlib.pyplot as plt

plt.close("all")

N = 100 # numărul de puncte dorite
x = np.linspace(0, 2*np.pi, N) # alege N valori de la 0 la 2*pi
noise = np.random.randn(N) # introducem niște erori (zgomot), distribuite normal
signal = 2*np.sin(x)

y = signal + noise # adăugăm erorile la calculul exact
plt.plot(x, signal) # graficul exact (continuu) = curba teoretică
plt.scatter(x, y) # graficul experimental, din puncte

plt.figure() # desenează pe o nouă figură; altfel, pune pe același grafic
y = signal + 0.2*noise # refacem desenul, cu erori mai mici cu 80%
plt.plot(x, signal)
plt.scatter(x, y)

plt.show()
````

Acum un exemplu de tip "pie chart".

````python
# pie
import matplotlib.pyplot as plt

plt.close("all")

x = [30, 20, 15, 25, 10] # lista de procentaje
dataName = ['data1', 'data2', 'data3', 'data4', 'data5']

plt.figure(figsize=(5,5)) # alegem mărimea figurii
plt.pie(x, labels=dataName) # desenează, luînd legenda din lista dataName

plt.show()
````

````python
# exploded pie
import matplotlib.pyplot as plt

plt.close("all")

x = [10, 10, 20, 20, 30]
dataName = ['data1', 'data2', 'data3', 'data4', 'data5']
explode = [0.01, 0, 1, 0.04, 0.009] # cît de mult să distanțeze feliile

plt.figure(figsize=(5,5))
plt.pie(x, explode=explode, labels=dataName, autopct='%.2f%%')
# autopct folosit pentru a afișa procentajele și numerele cu 2 zecimale
plt.show()
````

Mai jos, grafic de tip bare, cu mai multe valori comparative.

````python
# barchart
import numpy as np
import matplotlib.pyplot as plt

plt.close("all")

# prima figură
A = [2, 4, 8, 6] # valorile pentru Y
extra = [5, 6, 4, 1]
labels = [2005, 2010, 2015, 2020]

plt.subplot(1, 1, 1) # desenează, ținînd cont de posibile rescalări
plt.bar(labels, A, color='b', label='A') # punem și etichete
plt.bar(labels, extra, color='r', bottom=A, label='extra')
plt.legend() # afișăm și legenda

plt.show()

# a doua figură
x = [1, 2, 4]
y = [3.5, 3, 2]
z = [2, 3, 1.5]

width = 0.2
locs = np.arange(1, len(x) + 1) # returnează o listă de la 1 la len(x) + 1
plt.subplot(1, 1, 1)
plt.bar(locs, x, width=width, label='x')
plt.bar(locs + width, y, width=width, color="red", label='y')
plt.bar(locs + 2*width, z, width=width, color="black", label='z')
plt.legend()

plt.xticks([1.25, 2.25, 3.25], [2000, 2005, 2010]) # punctele de pe axa X

plt.show()
````
